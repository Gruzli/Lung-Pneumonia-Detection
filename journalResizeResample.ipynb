```python
import matplotlib.pyplot as plt
from PIL import Image

# Load the image
image_path = "2LabeledMask/img/1.jpeg"
original_image = Image.open(image_path)

# Resampling methods
resampling_methods = {
    "NEAREST": Image.NEAREST,
    "BILINEAR": Image.BILINEAR,
    "BICUBIC": Image.BICUBIC,
    "LANCZOS": Image.LANCZOS
}

# Resize images
resized_images = {}
for method_name, method in resampling_methods.items():
    target = 150
    resized_image = original_image.resize((target, target), method)
    resized_images[method_name] = resized_image

# Display images in a grid
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
for ax, (method_name, resized_image) in zip(axes.flatten(), resized_images.items()):
    ax.imshow(resized_image,cmap='gray')
    ax.set_title(method_name)
    ax.axis('off')

plt.tight_layout()
plt.show()
```
